```python
import tensorflow as tf
import numpy as np
```

```python
a_train = np.array([3, 4, 5, 6, 8, 10])
b_train = np.array([4, 5, 12, 8, 15, 24])
c_train = np.array([5, 7, 13, 10, 17, 26])
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(
    optimizer='adam', 
    loss='mean_squared_error'
    )
```

```python
print("Comienza Entrenamiento")
historial=model.fit(np.column_stack((a_train, b_train)), c_train, epochs=1000, verbose=False)
print("Termina Entrenamiento")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoc")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history["loss"])
```

```python
a_test = np.array([7, 9, 10])
b_test = np.array([24, 40, 18])
```

```python
c_pred = model.predict(np.column_stack((a_test, b_test)))
```

```python
for i in range(len(a_test)):
    print(f"Para a={a_test[i]}, b={b_test[i]}, c={c_pred[i]}")
```
